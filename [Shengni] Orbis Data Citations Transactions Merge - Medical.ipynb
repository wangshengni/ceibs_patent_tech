### Forward Citation

```python
open_pat_fwd_cit = open('D:/wsunny/Innovation/2020.7.10 BvD Orbis IP Datafeed For CEIBS/orbisIp_raw/Patents_ForwardCitations.txt',
         encoding='utf-8', errors='backslashreplace')
```

```python
pat_fwd_cit = pd.read_table(open_pat_fwd_cit,chunksize=1000000,error_bad_lines=False)
```

```python
%%time
# select PatPublNr in pat_fwd_cit that matches PatPublNr in join_pat_headers in order to reduce dataframe size

match_pat_fwd_cit = []
for chunks in pat_fwd_cit: 
    chunks.drop_duplicates(keep='first', inplace= True)
    xx = chunks.PatPublNr.isin(medeq.PatPublNr)
    yy = chunks[xx]
    match_pat_fwd_cit.append(yy)

match_pat_fwd_cit_concat = pd.concat(match_pat_fwd_cit)
```

```python
match_pat_fwd_cit_concat
```

```python
match_pat_fwd_cit_concat['PatPublNr'].nunique()
```

```python
join_pat_fwd_cit = medeq.merge(match_pat_fwd_cit_concat, how = 'left',on = 'PatPublNr')
join_pat_fwd_cit
```

### Backward Citation

```python
pat_bwd_cit = pd.read_table("D:/wsunny/Innovation/2020.7.10 BvD Orbis IP Datafeed For CEIBS/orbisIp_raw/Patents_BackwardCitations.txt", 
                            chunksize = 1000000)
```

```python
%%time

match_pat_bwd_cit = []
for chunks in pat_bwd_cit: 
    chunks.drop_duplicates(keep='first', inplace = True)
    jj = chunks.PatPublNr.isin(join_pat_fwd_cit.PatPublNr)
    kk = chunks[jj]
    match_pat_bwd_cit.append(kk)

match_pat_bwd_cit_concat = pd.concat(match_pat_bwd_cit)
```

```python
match_pat_bwd_cit_concat
```

```python
match_pat_bwd_cit_concat['PatPublNr'].nunique()
```

```python
join_pat_bwd_cit = join_pat_fwd_cit.merge(match_pat_bwd_cit_concat, how = 'left',on = 'PatPublNr')
join_pat_bwd_cit
```

```python
join_pat_bwd_cit['PatPublNr'].nunique()
```

```python
col_names_bwd = join_pat_bwd_cit.select_dtypes(include=['object']).columns
col_names_bwd_list = list(col_names_bwd)
col_names_bwd_list
```

```python
open_pat_headers = open('D:/wsunny/Innovation/2020.7.10 BvD Orbis IP Datafeed For CEIBS/orbisIp_raw/Patents_Headers.txt',
         encoding='utf-8', errors='backslashreplace')
```

```python
select_pat_headers = ["PatPublNr","Publn_ctry","GrantDate","Priority_date","Publn_date"]
```

```python
pat_headers = pd.read_table(open_pat_headers,chunksize=1000000,error_bad_lines=False,usecols = select_pat_headers )
```

```python
%%time

match_pat_headers = []
for chunks in pat_headers: 
    chunks.drop_duplicates(keep='first', inplace = True)
    m = chunks.PatPublNr.isin(join_pat_bwd_cit.PatPublNr)
    n = chunks[m]
    match_pat_headers.append(n)

match_pat_headers_concat = pd.concat(match_pat_headers)
```

```python
join_pat_headers = join_pat_bwd_cit.merge(match_pat_headers_concat, how = 'left',on = 'PatPublNr')
```
